```python
import itertools
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from scipy.stats import ttest_ind

import glob
import scipy
import re
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 12
cm = 1/2.54
fig_folder = './Figure_2_exports'
data_folder = '../all_data/Figure_2_data'
```

```python
all_data = pd.read_excel(data_folder+'/Supp_fig-mCherry_vs_deGFP.xlsx')
all_data
```

```python
all_data.value_counts(['mCherry_conc'])
```

```python
all_data.value_counts(['GFP_conc'])
```

```python
fig, ax = plt.subplots(figsize=[9*cm,9*cm])

sns.lineplot(data=all_data,x='GFP_conc',y='GFP_AU',hue='mCherry_conc',marker='o',palette='crest',ax=ax)
plt.xlabel('GFP plasmid conc. (ng/uL)')
plt.ylabel('GFP Fluorescence (AU)')

plt.legend(title='mCherry (ng/uL)', loc='lower right')
plt.tight_layout()

# plt.savefig(f'{fig_folder}/Supp_fig-GFP_vs_mCherry_expression.svg',format='svg',dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=[9*cm,9*cm])

sns.lineplot(data=all_data,x='mCherry_conc',y='mCherry_AU',hue='GFP_conc',marker='o',palette='flare',ax=ax)

plt.xlabel('mCherry plasmid conc. (ng/uL)')
plt.ylabel('mCherry Fluorescence (AU)')

plt.legend(title='GFP (ng/uL)', loc='upper left')
plt.tight_layout()
# plt.savefig(f'{fig_folder}/Supp_fig-mCherry_vs_GFP_expression.svg',format='svg',dpi=300)

plt.show()
```
